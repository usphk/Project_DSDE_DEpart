```python
from os import defpath
import pandas as pd
import numpy as np
import csv
df = pd.read_csv('/content/output.csv', on_bad_lines='skip',engine='python')
print (df.shape)
```

```python
half_count = len(df) / 2

df = df.dropna(thresh=half_count,axis=1) # Drop any column with more than 50% missing values
print (df.shape)
```

```python
threshold = 0.6
def get_duplicate_ratio(column):
    return column.value_counts(normalize=True).max()  # ค่าที่ซ้ำที่สุด / จำนวนข้อมูลทั้งหมด


duplicate_ratios = df.apply(get_duplicate_ratio)

cols_to_drop = duplicate_ratios[duplicate_ratios > threshold].index
df[cols_to_drop]


df = df.drop(columns=cols_to_drop)

print(df.shape)
```
